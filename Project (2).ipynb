```python
import matplotlib.pyplot as plt
Outlet_Type_counts=dataset['Outlet_Type'].value_counts()
plt.bar(Outlet_Type_counts.index,Outlet_Type_counts.values)
plt.xlabel('Outlet_Type')
plt.ylabel('Count')
plt.title('Count of items in Each Outlet_Type')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(dataset['Item_MRP'],bins=60) #bin is an attribute which decides the width of the graph #more the number lesser the width
plt.xlabel('Item_MRP') #value
plt.ylabel('Frequency') #frequency
plt.title('Distribution of Item_MRP ')
plt.show()
```

```python
Outlet_Identifier_counts=dataset['Outlet_Identifier'].value_counts()
plt.pie(Outlet_Identifier_counts,labels=Outlet_Identifier_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Outlet_Identifier')
plt.show
```
